```python
df["Average_Score"] = (df["math score"] + df["reading score"] + df["writing score"])/3
```

```python
print(df[df["test preparation course"] == "none"]["Average_Score"].mean())
print(df[df["test preparation course"] == "completed"]["Average_Score"].mean())
```

The Test Preparation Course does affect the marks.

The Average score of students who took the course is greater when compared to non test takers.

```python
print(df[df["lunch"] == "standard"]["Average_Score"].mean())
print(df[df["lunch"] == "free/reduced"]["Average_Score"].mean())
```

The Average score of students who took standard lunch is greater.
